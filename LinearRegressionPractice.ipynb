# **Univariate Linear Regression**

```python
import pandas as pd
import matplotlib.pyplot as plt
import csv
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv('House_Rent_Dataset.csv')
data.head()
```

```python
X = data['size'].values
Y = data['price'].values
```

```python
mean_X = np.mean(X)
mean_Y = np.mean(Y)
```

```python
A = 0
B = 0
for i in range(len(X)):
  A += (X[i] - mean_X) * (Y[i] - mean_Y)
  B += (X[i] - mean_X) ** 2

m = A / B
c = mean_Y - (m * mean_X)

print(m, c)
```

```python
plt.rcParams['figure.figsize'] = (10.0, 5.0)

max_x = np.max(X) + 100
min_x = np.min(X) - 100

x = np.linspace(min_x, max_x, 1000)
y = m * x + c

plt.plot(x, y, color='#58b700', label='Regression Line')
plt.scatter(X, Y, c = "#ef4456", label = 'Scatter Plot')

plt.xlabel('Area')
plt.ylabel('Price')
plt.legend()
plt.show()
```

# **Gradient Descent**

```python
import matplotlib.pyplot as plt

thetas = []
costs = []

m = len(X)
epoch = 100
learning_rate = 0.1
theta_0 = 0
theta_1 = 0.5425615384615384

for i in range(epoch):
    temp_0 = 0
    temp_1 = 0
    for j in range(m):
        temp_0 += (theta_0 + theta_1*X[j] - Y[j])
        temp_1 += (theta_0 + theta_1*X[j] - Y[j]) * X[j]

    theta_0 = theta_0 - learning_rate * (temp_0) / m
    theta_1 = theta_1 - learning_rate * (temp_1) / m  # Uncommented this line for consistency

    thetas.append(theta_0)

    # cost function
    sum_squared_errors = 0
    for k in range(m):
        sum_squared_errors += (theta_0 + theta_1*X[k] - Y[k]) ** 2
    cost = sum_squared_errors / (2 * m)
    costs.append(cost)

print('Final Thetas:', theta_0, theta_1)

plt.plot(range(epoch), costs, label='Cost Function')
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.legend()
plt.show()
```

# **Gradient Descent with all the data**

```python
data = pd.read_csv('4K_House_Rent_Dataset.csv')
data.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt

X = data['Size']
Y = data['Rent']

m = len(X)
epoch = 1000
learning_rate = 0.0000001
theta_0 = 0
theta_1 = 0

costs = []

for i in range(epoch):
    temp_0 = 0
    temp_1 = 0
    for j in range(m):
        temp_0 += (theta_0 + theta_1*X[j] - Y[j])
        temp_1 += (theta_0 + theta_1*X[j] - Y[j]) * X[j]

    theta_0 = theta_0 - learning_rate * (temp_0) / m
    theta_1 = theta_1 - learning_rate * (temp_1) / m

    # cost function
    sum_squared_errors = 0
    for k in range(m):
        sum_squared_errors += (theta_0 + theta_1*X[k] - Y[k]) ** 2
    cost = sum_squared_errors / (2 * m)
    costs.append(cost)
```

```python
plt.plot(range(epoch), costs, label='Cost Function')
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.legend()
plt.show()

max_x = np.max(X) + 100
min_x = np.min(X) - 100

x = np.linspace(min_x, max_x, 1000)
y = theta_1 * x + theta_0

plt.plot(x, y, color='#58b130', label='Regression Line')
plt.scatter(X, Y, c="#3498db", label='Scatter Plot')

print('Final Thetas:', theta_0, theta_1)

plt.xlabel('Area')
plt.ylabel('Price')
plt.legend()
plt.show()
```
